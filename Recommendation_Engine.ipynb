```python
# change structure of dataset so that we can compute the similarity score
user_movies_df = movies_df.pivot(index='userId', # row = userID
                                 columns='movie',  # Col = Movie name
                                 values='rating')
```

```python
#Impute those NaNs with 0 values ) 0 - did not watch movie
user_movies_df.fillna(0,inplace=True)
```

```python
user_movies_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
# User similarity - pairwise - taking 2 rows at a time it will compute distance
# That distance is scaled between 0 to 1
# Suppose distance between 2 rows is 0.  We want similarity, not a distance.
# That is calculated as Similarity=1-distance. So 1-0=1. i.e. similarity in rows is 1.
# Suppose distance is 0.9. So 1-0.9=0.1

user_sim = 1 - pairwise_distances(user_movies_df.values,metric='cosine')

# pairwise_distances(..., metric='cosine') calculates pairwise distances between rows (users) in this matrix using the cosine distance metric.
# 1 - pairwise_distances(...) converts distances to similarities.
# Cosine distance ranges from 0 to 2, where 0 indicates no distance and 2 indicates maximum distance,
# so subtracting from 1 inverts this to give cosine similarity which ranges from -1 to 1.
```

```python
user_sim   # This is similarity matrix - diagonal values are 1
```

```python
np.fill_diagonal(user_sim,0)
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
#Set the index and column names to user ids
user_sim_df.index = movies_df.userId.unique()
user_sim_df.columns = movies_df.userId.unique()
```

# **Show number of common movies**

```python
# Binary matrix of who rated what
binary_df = user_movies_df.map(lambda x: 1 if x > 0 else 0)

# Dot product of binary_df with its transpose = count of common ratings
overlap_matrix = binary_df.dot(binary_df.T)

# Check how many movies user 6 and 168 rated in common
overlap_matrix.loc[8, 16] # try 6 an, 168, 717 and 7117, 6 and 7117
```

# **Step-by-step to find users similar to userId = 7117:**

```python
# 1. Get similarity scores for userId 7117
similar_users_7117 = user_sim_df.loc[7117]

# 2. Sort descending to get most similar users first
similar_users_7117 = similar_users_7117.sort_values(ascending=False)

# 3. View top 5 similar users (excluding themselves)
print(similar_users_7117.head(5))  # or head(n)
```
